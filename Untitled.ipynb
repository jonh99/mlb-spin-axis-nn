```python
cut_train_df = train_df[['release_speed', 'release_pos_x', 'release_pos_z', 'spin_axis']].dropna()
```

```python
cut_test_df = test_df[['release_speed', 'release_pos_x','release_pos_z','spin_axis']].dropna()
```

```python
# Calculate the Z-scores of each column in the training set:
cut_train_df_mean = cut_train_df.mean()
cut_train_df_std = cut_train_df.std()
cut_train_df_norm = (cut_train_df - cut_train_df_mean)/cut_train_df_std
```

```python
# Calculate the Z-scores of each column in the test set.
cut_test_df_mean = cut_test_df.mean()
cut_test_df_std = cut_test_df.std()
cut_test_df_norm = (cut_test_df - cut_test_df_mean)/cut_test_df_std

cut_test_df_norm.head()
```

```python
# Create an empty list that will eventually hold all created feature columns.
feature_columns = []
```

```python
# We scaled all the columns, including latitude and longitude, into their
# Z scores. So, instead of picking a resolution in degrees, we're going
# to use resolution_in_Zs.  A resolution_in_Zs of 1 corresponds to 
# a full standard deviation. 
resolution_in_Zs = 0.3  # 3/10 of a standard deviation.
```

```python
# Create a bucket feature column for release position x.
release_pos_x_as_a_numeric_column = tf.feature_column.numeric_column("release_pos_x")
release_pos_x_boundaries = list(np.arange(int(min(cut_train_df_norm['release_pos_x'])), 
                                     int(max(cut_train_df_norm['release_pos_x'])), 
                                     resolution_in_Zs))
release_pos_x = tf.feature_column.bucketized_column(release_pos_x_as_a_numeric_column, release_pos_x_boundaries)
```

```python
# Create a bucket feature column for release position z.
release_pos_z_as_a_numeric_column = tf.feature_column.numeric_column("release_pos_z")
release_pos_z_boundaries = list(np.arange(int(min(cut_train_df_norm['release_pos_z'])), 
                                      int(max(cut_train_df_norm['release_pos_z'])), 
                                      resolution_in_Zs))
release_pos_z = tf.feature_column.bucketized_column(release_pos_z_as_a_numeric_column, 
                                                release_pos_z_boundaries)
```

```python
# Create a feature cross of release pos x and y
release_x_by_release_z = tf.feature_column.crossed_column([release_pos_x, release_pos_z], hash_bucket_size=100)
crossed_feature = tf.feature_column.indicator_column(release_x_by_release_z)
feature_columns.append(crossed_feature) 
```

```python
# Represent release speed as a floating-point value.
release_speed = tf.feature_column.numeric_column("release_speed")
feature_columns.append(release_speed)
```

```python
# Convert the list of feature columns into a layer that will later be fed into
# the model. 
my_feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
```

```python
def create_model(my_learning_rate, my_feature_layer):
  """Create and compile a simple linear regression model."""
  # Most simple tf.keras models are sequential.
  model = tf.keras.models.Sequential()

  # Add the layer containing the feature columns to the model.
  model.add(my_feature_layer)

  # Describe the topography of the model by calling the tf.keras.layers.Dense
  # method once for each layer. We've specified the following arguments:
  #   * units specifies the number of nodes in this layer.
  #   * activation specifies the activation function (Rectified Linear Unit).
  #   * name is just a string that can be useful when debugging.

  # Define the first hidden layer with 20 nodes.   
  model.add(tf.keras.layers.Dense(units=20, 
                                  activation='relu', 
                                  name='Hidden1'))
  
  # Define the second hidden layer with 12 nodes. 
  model.add(tf.keras.layers.Dense(units=12, 
                                  activation='relu', 
                                  name='Hidden2'))
  
  # Define the output layer.
  model.add(tf.keras.layers.Dense(units=1,  
                                  name='Output'))                              
  
  model.compile(optimizer=tf.keras.optimizers.Adam(lr=my_learning_rate),
                loss="mean_squared_error",
                metrics=[tf.keras.metrics.MeanSquaredError()])

  return model
```

```python
def train_model(model, dataset, epochs, label_name,
                batch_size=None):
  """Train the model by feeding it data."""

  # Split the dataset into features and label.
  features = {name:np.array(value) for name, value in dataset.items()}
  label = np.array(features.pop(label_name))
  history = model.fit(x=features, y=label, batch_size=batch_size,
                      epochs=epochs, shuffle=True) 

  # The list of epochs is stored separately from the rest of history.
  epochs = history.epoch
  
  # To track the progression of training, gather a snapshot
  # of the model's mean squared error at each epoch. 
  hist = pd.DataFrame(history.history)
  mse = hist["mean_squared_error"]

  return epochs, mse
```

```python
# The following variables are the hyperparameters.
learning_rate = 0.01
epochs = 20
batch_size = 1000

# Specify the label
label_name = "spin_axis"


# Establish the model's topography.
my_model = create_model(learning_rate, my_feature_layer)
```

```python
# Train the model on the normalized training set. We're passing the entire
# normalized training set, but the model will only use the features
# defined by the feature_layer.
epochs, mse = train_model(my_model, cut_train_df_norm, epochs, 
                          label_name, batch_size)
plot_the_loss_curve(epochs, mse)
```
